```python
# --- 驗證程式碼 ---
common_keys = set(common_codes.keys())
missing_common_codes_vars = []
incorrect_common_code_values = []

# # 使用繁體中文註解
# 迭代檢查每個變項的 value map
for var_name, var_map in value_maps_merged.items():
    # 確保 var_map 是字典類型
    if not isinstance(var_map, dict):
        print(f"警告：變項 '{var_name}' 的值不是字典，跳過檢查。")
        continue

    current_keys = set(var_map.keys())

    # 檢查是否包含所有 common_codes 的鍵
    # 注意：這裡假設所有通用碼原則上都應存在於每個 map 中
    missing_keys = common_keys - current_keys
    if missing_keys:
        missing_common_codes_vars.append((var_name, list(missing_keys)))

    # 檢查已存在的通用碼的值是否正確（或被特定值覆蓋）
    for code in common_keys:
        if code in current_keys:
            # 如果這個 code 在通用碼中，但在特定定義中沒有這個 code（或說明相同），
            # 那麼它的值應該等於通用碼的值。
            # 這個檢查比較複雜，因為我們沒有原始的 base map 來直接比對。
            # 一個簡化的檢查是：如果 value_maps_merged[var_name][code] 的值
            # 既不是 common_codes[code] 的值，也不是該變項原本預期的特定值，
            # 那就可能有問題。但這個 "原本預期的特定值" 不易直接判斷。
            # 因此，這裡主要依賴合併邏輯 {**common_codes, **specific_map} 的正確性。
            # 我們可以抽查幾個特定覆蓋的例子，例如 -7。
            if code == -7 and var_name == 'as3' and var_map[code] != '我不清楚':
                 incorrect_common_code_values.append((var_name, code, var_map[code], "預期 '我不清楚'"))
            # 可以為其他已知特定覆蓋的代碼添加類似檢查

# --- 報告結果 ---
if not missing_common_codes_vars:
    print("初步檢查：所有變項的 value map 似乎都包含了通用代碼的鍵。")
else:
    print("警告：以下變項的 value map 缺少部分通用代碼鍵：")
    for var, missing in missing_common_codes_vars:
        print(f"- {var}: 缺少 {missing}")

if incorrect_common_code_values:
     print("\n警告：發現部分通用代碼的值與預期不符（檢查特定覆蓋情況）：")
     for var, code, actual, expected in incorrect_common_code_values:
         print(f"- {var}: 代碼 {code} 的值是 '{actual}', 但 {expected}")

print("\n--- 驗證說明 ---")
print("1. 上述程式主要檢查每個變項的 value map 是否『包含』了所有通用代碼的鍵。")
print("2. 對於通用代碼的『值』是否正確（特別是當有特定定義覆蓋通用定義時），建議搭配手動抽樣檢查。")
print("3. 如果您發現有變項缺少通用代碼鍵，請檢查生成 `value_maps_merged` 的過程。")
```

```python
# -*- coding: utf-8 -*-
import json
import pandas as pd
import pprint

# --- 步驟 1: 從 JSON 檔案讀取 value_maps ---
# *** 請確保這個 JSON 檔案是您更新過的、包含正確 as70a 映射的版本 ***
file_path = 'tigps_w1_value_maps.json'
loaded_value_maps = None

try:
    with open(file_path, 'r', encoding='utf-8') as f:
        loaded_value_maps = json.load(f)
    print(f"成功從 {file_path} 讀取 value_maps。")
except FileNotFoundError:
    print(f"錯誤：找不到檔案 {file_path}。請確認檔案路徑和名稱。")
    exit()
except json.JSONDecodeError as e:
    print(f"讀取或解析 JSON 檔案時發生錯誤： {e}")
    exit()
except Exception as e:
    print(f"發生預期外的錯誤：{e}")
    exit()

# --- 步驟 2: 處理內部字典的鍵，從字串轉回整數 ---
processed_value_maps = {}
if loaded_value_maps:
    for var_name, inner_map in loaded_value_maps.items():
        if isinstance(inner_map, dict):
            new_inner_map = {}
            for key_str, value in inner_map.items():
                try:
                    key_int = int(key_str)
                    new_inner_map[key_int] = value
                except ValueError:
                    new_inner_map[key_str] = value
            processed_value_maps[var_name] = new_inner_map
        else:
            processed_value_maps[var_name] = inner_map
    print("\n已將 value_maps 內部字典的鍵嘗試轉換回整數。")
else:
    print("錯誤：未能成功載入 value_maps。")
    exit() # 如果沒有 map，無法繼續

# --- 步驟 3: 載入您的實際資料 ---
# *** 請確保您的 CSV 文件路徑正確 ***
csv_file_path = '../data/TIGPSw1_s.csv'
raw_data_df = None

try:
    raw_data_df = pd.read_csv(csv_file_path)
    print(f"\n成功從 {csv_file_path} 載入資料。")
except FileNotFoundError:
    print(f"錯誤：找不到您的資料檔案 {csv_file_path}。請確認路徑。")
    exit()
except Exception as e:
    print(f"載入 CSV 檔案時發生錯誤：{e}")
    exit()

# --- 步驟 4: 選取前五筆資料進行測試 ---
df_to_test = raw_data_df.head().copy() # 使用 .copy() 避免 SettingWithCopyWarning

print("\n--- 您實際資料的前五筆 (raw_data_df.head()) ---")
print(df_to_test)

# --- 步驟 5: 對選擇的欄位進行轉換 ---
# 可以再加上一些其他的，例如 'as4e1' (第一照顧者健康)
columns_to_map = ['asl', 'as3', 'as13', 'as62', 'as70a', 'as4e1', 'as70b1', 'as70b2']

print("\n--- 開始進行轉換 ---")
for col in columns_to_map:
    label_col = f"{col}_label" # 新增的標籤欄位名稱

    # 檢查欄位是否存在於 DataFrame 中
    if col not in df_to_test.columns:
        print(f"警告：欄位 '{col}' 不在您的 DataFrame 中，跳過轉換。")
        continue

    # 檢查欄位是否存在於處理後的 value map 中
    if col in processed_value_maps:
        value_map = processed_value_maps[col]
        # 執行轉換
        # 注意：如果 df_to_test[col] 中有 processed_value_maps[col] 沒有的鍵，結果會是 NaN
        df_to_test[label_col] = df_to_test[col].map(value_map)
        print(f"已轉換欄位: '{col}' -> '{label_col}'")
    else:
        print(f"警告：在 value_maps 中找不到欄位 '{col}' 的對應，跳過轉換。")

print("\n--- 實際資料前五筆轉換結果 (包含標籤欄位) ---")
pd.set_option('display.max_columns', None) # 顯示所有欄位
pd.set_option('display.width', 1000) # 加寬顯示寬度
print(df_to_test)

print("\n--- 說明 ---")
print("1. 上方顯示的是您 CSV 檔案前五筆數據，以及對選定欄位應用 value map 轉換後的結果。")
print(f"2. 轉換後的文字標籤儲存在以 '_label' 結尾的新欄位中。")
print(f"3. 如果原始欄位的值在對應的 value map 中找不到（例如資料輸入錯誤或未定義的代碼），轉換結果會顯示為 NaN。")

# --- 關於 descriptive_df ---
# 如果您之前執行過 descriptive_df = raw_data_df.rename(columns=loaded_id_map)
# descriptive_df 的 *欄位名稱* 是中文的，但裡面的 *值* 仍然是數字代碼。
# 因此，value map 的轉換仍然是應用在這些包含數字代碼的欄位上，
# 只是您選取欄位時要用中文名稱（如果 descriptive_df 存在）。
# 例如: descriptive_df['性別_label'] = descriptive_df['請問你的性別(生理性別)?'].map(processed_value_maps['asl'])
# 但通常建議直接在原始代碼欄位 (raw_data_df['asl']) 上操作，更容易管理。
```
